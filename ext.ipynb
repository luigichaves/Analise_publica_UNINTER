### Análise e criação de gráficos

Contagem de órgãos que mais firmam acordo

```python
tf_rec['nome_orgao_tra'].value_counts() # contagem de orgaos que mais firmam acordos
```

```python
contagem_orgaos = tf_rec['nome_orgao_tra'].value_counts().head(6)

plt.figure(figsize=(10, 6))
sns.barplot(x=contagem_orgaos.values, y=contagem_orgaos.index, palette="viridis")

plt.title('6 ÓRGÃOS QUE MAIS FIRMAM ACORDOS')
plt.xlabel('QUANTIDADE DE ACORDOS')
plt.ylabel('ÓRGÃOS')

#plt.savefig('1.png', dpi=300, bbox_inches='tight')

plt.show()
```

Quantidade de acordos firmados por ano (levando em conta somente os 4 maiores)

```python
df4 = tf_rec.loc[(tf_rec['nome_orgao_tra'] == 'Ministério da Saúde - Unidades com vínculo direto') |
                (tf_rec['nome_orgao_tra'] == 'Sem informação') |
                (tf_rec['nome_orgao_tra'] == 'Fundo Nacional de Desenvolvimento da Educação') |
                (tf_rec['nome_orgao_tra'] == 'Fundo Nacional de Assistência Social')].copy()
```

```python
df4['nome_orgao_tra'].value_counts()
```

```python
df4['ano'] = df4['ano_mes_tra'].dt.year # criando a coluna ano
```

```python
import warnings

warnings.filterwarnings('ignore') # desativando warnings

contagem_orgaos_por_ano = df4.groupby(['ano', 'nome_orgao_tra']).size().reset_index(name='contagem')

sns.set(style="whitegrid")
sns.set_palette("Set1")

plt.figure(figsize=(10, 6))
sns.lineplot(x='ano', y='contagem', hue='nome_orgao_tra', data=contagem_orgaos_por_ano, marker='o')
plt.xticks([2020, 2021, 2022, 2023, 2024])

plt.title('QUANTIDADE DE ACORDOS POR ÓRGÃO')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Órgãos')
plt.grid(True)
plt.legend(title='Nome do Órgão', bbox_to_anchor=(1.05, 1), loc='upper left')

#plt.savefig('2.png', dpi=300, bbox_inches='tight')

plt.show()
```
